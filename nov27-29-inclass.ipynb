```python
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

#correlogram
sns.pairplot(df)
sns.plt.show()
```

```python
#line of best fit
sns.pairplot(df, kind="reg")
plt.show()

sns.pairplot(df, kind="scatter")
plt.show()
```

```python
# left
sns.pairplot(df, kind="reg", hue="species", markers=["o", "s", "D"], palette="Set2")
plt.show()
 
# right: you can give other arguments with plot_kws.
sns.pairplot(df, kind="reg", hue="species", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```
