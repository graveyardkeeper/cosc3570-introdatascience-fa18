```python
import seaborn as sns
df = sns.load_dataset('iris')
import matplotlib.pyplot as plt
#importing pandas
import pandas as pd
#read in the housing data
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
from scipy import stats
from scipy.stats import norm
df = pd.read_csv('realestate.csv')
df
```

```python
realestate = df.rename(index=str, columns={"X1 transaction date": "Transaction Date", "X2 house age": "House Age", "X3 distance to the nearest MRT station": "Distance to Mass Rapid Transportation", "X4 number of convenience stores": "Number of convenience stores", "X5 latitude": "Latitude", "X6 longitude": "Longitude", "Y house price of unit area": "House Price of Unit Area"})
```

```python
realestate
```

```python
realestate1 = realestate.drop(columns=['Latitude', 'Longitude', 'Transaction Date'])
```

```python
realestate1
```

```python
sns.distplot(realestate1["House Price of Unit Area"])

#Calculate the Kolmogorov–Smirnov statistic to test for normality
houseprice = realestate1["House Price of Unit Area"]
stats.kstest(houseprice, 'norm', norm.fit(realestate1["House Price of Unit Area"]))
```

```python
realestate2 = realestate1.drop(columns=['No'])
realestate2.describe()
```

```python
sns.distplot(realestate1["House Age"])
houseprice = realestate1["House Age"]
stats.kstest(houseprice, 'norm', norm.fit(realestate1["House Age"]))
```
